```python
from deepdrive_hardware.deepdrive_hardware.pwm_motor import MotorDriverL293
import time

robot = MotorDriverL293()

robot.set_motors(20, 20)
time.sleep(1)

robot.set_motors(100, 100)
time.sleep(1)

robot.set_motors(100, 0)
time.sleep(1)

robot.set_motors(0, 100)
time.sleep(1)

robot.set_motors(0, 0)


for i in range(100):
    robot.set_motors(i, i)
    time.sleep(0.1)

for i in range(100, 0, -1):
    robot.set_motors(i, i)
    time.sleep(0.1)


for i in range(0, 100, 10):
    robot.set_motors(i, 0)
    time.sleep(0.1)

for i in range(100, 0, -10):
    robot.set_motors(i, 0)
    time.sleep(0.1)

robot.set_motors(0, 0)

for i in range(0, 100, 10):
    robot.set_motors(0, i)
    time.sleep(0.1)

for i in range(100, 0, -10):
    robot.set_motors(0, i)
    time.sleep(0.1)

robot.set_motors(0, 0)
```
